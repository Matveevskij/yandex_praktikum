## Предобработка данных

###  Обработка пропусков

Посчитаем, сколько в таблице пропущенных значений.

```python
df.isna().sum()
```

Выведем на экран значения, которые показывают % пропусков в каждом столбце DataFrame

```python
pd.DataFrame(round(df.isna().mean()*100,)).style.background_gradient('coolwarm')
```

Первым делом, нужно удалить пропуски в столбце `locality_name`(название населённого пункта). Т.к. заменить их не получится.

```python
df = (df.dropna(subset=['locality_name'])).reset_index(drop=True)
```

Удалим строки с долей пропусков менее 10 %

```python
df = (df.dropna(subset=['locality_name'])).reset_index(drop=True)
```

```python
df = (df.dropna(subset=['floors_total'])).reset_index(drop=True)
```

Если продавец не указал число балконов, то скорее всего их нет. Правильнее будет заменить на 0

```python
df['balcony'] = df['balcony'].fillna(0)
```

Пропуски в длительности объявления вероятнее появились из-за технической ошибки. Будет правильнее удалить эти пропуски.

Столбец `is_apartment` имеет булевый тип. В первоначальном DataFrame в этом столбце имеются только 'True'. Исходя из этого, можно сделать вывод, что продавец не заполнил эту графу, т.к. его недвижимость не является аппартаментом. Зполнил пропуски 'False'.

```python
df['is_apartment'] = df['is_apartment'].fillna(False)
```

Проверим значения высоты потолков, на наличие аномальных значений:

```python
df[df['ceiling_height']>=20]
```

Выяснили, что в DF имется недвижимость со 100 метровыми потолками. Возможно кто-то считал в дециметрах, переведем в стандартные метры:

```python
def correct_сeiling_height(i): #функция, для перевода в метры
    if i >= 20:
        i = i/10
    return i
df['ceiling_height'] = df['ceiling_height'].apply(correct_сeiling_height) # заменим на нормальные значения
```

Заполним пропуски в высоте потолков в зависимости от населенного пункта (в каждом районе строят квартиры с одниковыми потолками, поэтому считаю, что нужно заменить пропуски таким образом).

```python
for locality_name in df['locality_name'].unique():
    median = df.loc[df['locality_name'] == locality_name,'ceiling_height'].median()
    df.loc[(df['ceiling_height'].isna()) & (df['locality_name'] == locality_name), 'ceiling_height'] = median

# все что не заполнилось заполняем общей медианой по Лен. области (121 пропуск заменим медианой)
df.loc[df['ceiling_height'].isna(), 'ceiling_height'] = df['ceiling_height'].median()
```

Заполняем пропуски в расстоянии до центра в зависимости от населенного пункта. Если останутся пропуски, то будет некорректно до-заполнять медианой).

```python
for locality_name in df['locality_name'].unique():
    median = df.loc[df['locality_name'] == locality_name,'cityCenters_nearest'].median()
    df.loc[(df['cityCenters_nearest'].isna()) & (df['locality_name'] == locality_name), 'cityCenters_nearest'] = median
```

Проверим столбец rooms:

```python
df['rooms'].value_counts()
```

Есть строки, где 0 комнат. Скорее всего это обычная студия, и переведем в 1-комнатную недвижимость.

```python
df['rooms'] = df['rooms'].replace(0,1)
```

Проверим наличие дубликатов:

```python
df.duplicated().sum()
```

Вывод: Удалили строки, где % пропусков в столбцах был менее 10%. Заменили отсутствующие значения на медианные в зависимости от населенного пунка (Ведь одном районе, строят одинаковые дома. Соответственно и некоторые параметры квартир - одинаковы). Так же в столбце `ceiling_height` заменили "очень высокие" потолки (20+ метров) на адекватные значения. 

Для некоторых типов данных нет подходящих значений под замену. В этом случае, правильнее оставить эти значения пустыми, чем запонять медианой.

```python
def number_of_floors (i):
    floors_total = i['floors_total']
    floor=i['floor']
    if floor == 1: 
        return 'первый'
    if floor == floors_total:
        return 'последний'
    if 1 < floor < floors_total: 
        return 'другой'
    
# Создадим новый столбец, который показывает варианта этажа (первый, последний или другой)
df['level'] = df.apply(number_of_floors, axis = 1)
```

```python
def column_visible(df, column, bins, range1, suptitle, xlabel, ylabel):

    plt.figure(figsize=(40,5)) # размер фигуры
    plt.suptitle(suptitle + ' [' + column + ']', fontsize=16) # название фигуры

    # первая фигура hist()
    plt.subplot(1, 2, 1) # выделим место в фигуре
    plt.hist(df[column], bins=bins, range = range1)
    plt.xlabel(xlabel) # название оси x
    plt.ylabel(ylabel) # название оси y
    plt.grid(axis = 'both') # отображение сетки
    
    
    # вывод фигуры
    plt.show()
```

```python
def yashik_s_ysami(df, column, bins, suptitle, xlabel):

    plt.figure(figsize=(16,5)) # размер фигуры
    plt.suptitle(suptitle + ' [' + column + ']', fontsize=16) # название фигуры


    # вторая фигура boxplot()
    plt.subplot(1, 2, 2) # выделим место в фигуре
    plt.boxplot(df[column], vert=False)
    plt.xlabel(xlabel) # название оси x
    plt.grid(axis = 'both') # отображение сетки

    # вывод фигуры
    plt.show()
```
